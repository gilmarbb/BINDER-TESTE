```python
import matplotlib.pyplot as plt
import numpy as np

# Parâmetros para desenhar a cicloide
r = 1  # raio do círculo
t = np.linspace(0, 2*np.pi, 100)  # parâmetro t variando de 0 a 2π

# Equações paramétricas da cicloide
x = r * (t - np.sin(t))
y = r * (1 - np.cos(t))

# Criando o gráfico
plt.figure(figsize=(10, 4))
plt.plot(x, y, label='Cicloide')
plt.title('Gráfico de uma Cicloide')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Parâmetros
r = 1  # raio do círculo
num_frames = 100  # número de quadros para a animação
t = np.linspace(0, 4*np.pi, num_frames)  # parâmetro t para a cicloide

# Função para calcular a cicloide
def calculate_cycloid(t):
    x = r * (t - np.sin(t))
    y = r * (1 - np.cos(t))
    return x, y

# Configuração da figura
fig, ax = plt.subplots()
ax.set_xlim(0, 4*np.pi*r)
ax.set_ylim(-r, 2*r)
line, = ax.plot([], [], color='blue')

# Função de inicialização
def init():
    line.set_data([], [])
    return line,

# Função para atualizar cada frame
def update(frame):
    x, y = calculate_cycloid(t[:frame])
    line.set_data(x, y)
    return line,

# Criando a animação
ani = FuncAnimation(fig, update, frames=num_frames, init_func=init, blit=True, interval=50, repeat=False)

plt.show()
```
